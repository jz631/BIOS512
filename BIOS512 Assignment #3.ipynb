```r
df = read.csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/bob-ross/elements-by-episode.csv')
df %>% head
```

```r
df1 = df %>%
    pivot_longer(cols = -one_of('EPISODE','TITLE'),names_to='ele',values_to='x') %>%
    filter(x>0) %>%
    mutate(ele = fct_lump(ele,n=10) %>% fct_infreq() %>% fct_rev())
df1 %>% head
```

```r
options(repr.plot.width = 4, repr.plot.height = 5, repr.plot.res = 200)
p = ggplot(df1, aes(x=ele)) + geom_bar(fill="#3182bd") 
p = p + coord_flip(ylim=c(0,1500))
p = p + geom_text(stat='count',aes(label = ..count..),hjust = -.1)
p = p + labs(title = "The Paintings of Bob Ross", subtitle = "count containing each element",x=" ", y  = " ")
p = p + theme(axis.text.x = element_blank(),panel.grid.major = element_blank(),panel.grid.minor = element_blank(),axis.ticks = element_blank())
p
```
